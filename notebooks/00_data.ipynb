```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/enma/Projects/Covid19/data/time_series_covid19_recovered_global.csv')
df.head()
```

```python
df.info()
```

```python
id_cols = ["Province/State", "Country/Region", "Lat", "Long"]
df_melted = df.melt(id_vars=id_cols, var_name="Date", value_name="Recovered")
```

```python
df_melted.head()
```

```python
df_melted = df_melted.rename(columns={
    "Province/State": "Province",
    "Country/Region": "Country"
})
```

```python
df_melted.head()
```

```python
df_melted["Date"] = pd.to_datetime(df_melted["Date"], format="%m/%d/%y")
```

```python
df_melted.dtypes
```

```python
df_melted = df_melted.drop(columns=["Lat","Long","Province"])
```

```python
df_melted.head()
```

```python
print(df_melted.shape)

print(df_melted.isna().sum())
print(df_melted["Country"].nunique())
print(df_melted["Date"].min(), df_melted["Date"].max())
```

```python
df_melted.to_csv(
    "/home/enma/Projects/Covid19/data/processed/covid_confirmed_recovered.csv",
    index=False
)
```
